```python
# adding the missing regions

country_region_mapping = {
    'Taiwan Province of China': 'East Asia',
    'Belize': 'Latin America and Caribbean',
    'Hong Kong S.A.R., China': 'East Asia',
    'Somalia': 'Sub-Saharan Africa',
    'Namibia': 'Sub-Saharan Africa',
    'South Sudan': 'Sub-Saharan Africa',
    'Trinidad & Tobago': 'Latin America and Caribbean',
    'Northern Cyprus': 'Northern Cyprus', 
    'North Macedonia': 'Southeastern Europe',
    'Gambia': 'Sub-Saharan Africa'
}
```

```python
for df in [df_2017, df_2018, df_2019]:
    mask = df['Region'].isnull() | (df['Region'] == 'Unknown')
    df.loc[mask, 'Region'] = df.loc[mask, 'Country'].map(country_region_mapping)
```

#### 4 Preparing the Data to merge

```python
# renaming the columns to reflect the year they are from to make it easier for merging the data

# for df_2015
df_2015 = df_2015.rename(columns=lambda x: f"{x} 2015" if x not in ['Country', 'Region'] else x)

# for df_2016
df_2016 = df_2016.rename(columns=lambda x: f"{x} 2016" if x not in ['Country', 'Region'] else x)

# For df_2017
df_2017 = df_2017.rename(columns=lambda x: f"{x} 2017" if x not in ['Country', 'Region'] else x)

# For df_2018
df_2018 = df_2018.rename(columns=lambda x: f"{x} 2018" if x not in ['Country', 'Region'] else x)

# For df_2019
df_2019 = df_2019.rename(columns=lambda x: f"{x} 2019" if x not in ['Country', 'Region'] else x)
```

#### 5 Merging Data

```python
merged_df = df_2015

# List all other DataFrames in order
dfs = [df_2016, df_2017, df_2018, df_2019]

# Merge each subsequent DataFrame
for df in dfs:
    merged_df = pd.merge(merged_df, df, on=['Country', 'Region'], how='outer')
```
