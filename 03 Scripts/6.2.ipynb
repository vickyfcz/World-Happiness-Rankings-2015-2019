#### How do Health (Life Expectancy) and freedom relate to happiness?

```python
# Select only numeric columns for correlation analysis
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate correlation matrix for numeric variables only
corr_matrix = numeric_df.corr()

# Plot heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation of Key Variables (2015-2019)')
plt.show()
```

**Happiness Rank & Happiness Score:**

-0.01 (essentially no correlation). Since rank is inverted (lower rank = higher happiness), a near-zero positive correlation indicates rank and score are related but not perfectly aligned in the data or with some variability.  

**Happiness Score & Social Support:**

-0.47 (moderate negative). Since higher happiness scores mean better conditions, and social support is positive, the negative sign likely arises because higher social support correlates with lower happiness rank (since a lower number in rank means higher happiness).  

**Social Support & Family:**

-0.31 (moderate inverse). Higher family support is associated with higher social support, which in turn relates positively to happiness.

**Health (Life Expectancy) & Happiness Score:**

0.71 (strong positive). Better health correlates strongly with higher happiness.

**Freedom & Happiness Score:**

0.62 (strong positive). More perceived freedom is associated with higher happiness.

**Trust in Government & Happiness:**

0.57 with happiness-related variables, indicating trust correlates strongly with well-being.

**Generosity & Happiness:**

0.68 (strong positive). Countries with higher generosity tend to have higher happiness.

**Dystopia Residual & Happiness Score:**

-0.18 (weak negative). Higher residuals (unexplained dissatisfaction) slightly relate to lower happiness.

Variables like Health, Freedom, Trust, and Generosity show strong positive relationships with happiness.
Social Support has a moderate negative coefficient with rank (or positive with happiness), indicating support is vital for positive well-being.
These suggest that policies fostering health, freedom, social trust, and generosity can significantly influence happiness levels.

#### 4. Scatterplot

```python
# Filter for year 2015
df_2019 = df[df['Year'] == 2019]

sns.set(style='whitegrid')

# Happiness vs. Income (GDP per Capita) for 2019
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x='Economy (GDP per Capita)',
    y='Happiness Score',
    data=df_2019,
    hue='Region',
    palette='viridis'
)
plt.title('Happiness vs. Income (GDP per Capita) - 2019')
plt.xlabel('Economy (GDP per Capita)')
plt.ylabel('Happiness Score')
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

The scatterplot depicts a strong positive relationship between GDP per capita and happiness scores in 2019, indicating that wealthier countries tend to report higher levels of well-being. The data shows clear regional clustering: high-income regions like Western Europe, North America, and Australia & New Zealand are grouped at the upper end of both income and happiness, while regions such as Sub-Saharan Africa, the Middle East, and parts of Asia are clustered at lower levels. Despite the overall trend, some countries with relatively low GDP still maintain relatively high happiness scores, suggesting that factors beyond incomesuch as social support, health, and culturealso significantly influence well-being. Overall, the plot highlights that economic prosperity is a key driver of happiness, but regional differences and outliers emphasize the multifaceted nature of well-being, where income is an important but not exclusive factor.

#### 6. Categorical Plot

```python
# Aggregate average happiness score by region for 2019
avg_happiness_by_region_2019 = df_2019.groupby('Region')['Happiness Score'].mean().reset_index()

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Happiness Score', y='Region', data=avg_happiness_by_region_2019, palette='viridis')
plt.title('Average Happiness Score by Region (2019)')
plt.xlabel('Average Happiness Score')
plt.ylabel('Region')
plt.show()
```

This bar chart reveals regional disparities in average happiness scores in 2019. Western Europe leads with the highest average happiness, affirming its reputation for high well-being, driven by strong social support, wealth, and social cohesion. Australia and New Zealand also rank highly, reflecting similar positive socio-economic factors. Conversely, regions like Sub-Saharan Africa and South Asia show lower average happiness, indicating challenges such as weaker social support networks, lower income levels, and health disparities. These regional differences highlight how factors like economic development, social trust, healthcare, and support systems collaboratively influence overall happiness. The data underscores the importance of social and economic policies tailored to regional contexts to improve well-being globally.
